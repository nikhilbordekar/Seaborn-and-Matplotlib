# **SEABORN**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset('iris')
planets = sns.load_dataset('planets')
```

```python
tips
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips)
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue = 'day')
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue = 'day', size='size')
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue = 'day', size='size', palette='YlGnBu')
```

# **HISTOGRAM**

```python
sns.histplot(tips['tip'], bins=15)
```

```python
sns.histplot(tips['tip'], bins=15, kde=True)
```

# **DISTPLOT**

```python
sns.distplot(tips['tip'], bins=15, kde=True)
```

# **BARPLOT**

```python
sns.barplot(x='sex',y='tip', data=tips, palette='YlGnBu')
```

# **BOXPLOT**

```python
sns.boxplot(x='day',y='tip', data=tips, palette='YlGnBu')
```

# **STRIPPLOT**

```python
sns.stripplot(x='day', y='tip', data=tips, hue='sex', palette='YlGnBu')
```

```python
sns.stripplot(x='day', y='tip', data=tips, hue='sex', palette='YlGnBu', dodge=True)
```

# **JOIN PLOT: SCATTER PLOT + DISTRIBUTION PLOT**

```python
sns.jointplot(x='tip', y='total_bill', data=tips)
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='reg')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='kde')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='kde', shade=True)
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='kde', shade=True, cmap='YlGnBu') #cmap=color_map
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='hex',  cmap='YlGnBu') #kind=hex and no shade
```

# **PAIRPLOT**

```python
sns.pairplot(titanic.select_dtypes(['number']), hue='pclass')
```

# **HEATMAP**

```python
sns.heatmap(titanic.corr(), cmap='YlGnBu')
```

```python
sns.heatmap(titanic.corr(), cmap='YlGnBu', annot=True) #cmap= coolwarm, icefire
```

# **CLUSTERMAP**

```python
sns.clustermap(iris.drop(['species'], axis=1))
```

# **SCATTER PLOT**

```python
x = np.random.random(50)
y = np.random.random(50)

plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y, marker='+')
plt.show()
```

```python
plt.scatter(x,y, marker='+', c='#00f', s=150) #size and color
plt.show()
```

```python
x = np.random.random(1000)
y = np.random.random(1000)
plt.scatter(x,y, marker='+', c='#00f', s=150, alpha=0.3) #transperancy
plt.show()
```

```python
years = [2006 + x for x in range(16) ]
weights = [81, 83, 86, 83,88, 90, 78, 66, 89, 100, 87,77, 133, 76, 98, 79]
```

```python
plt.plot(years, weights, c='r')
```

```python
plt.plot(years, weights, c='r', lw=3)
```

```python
plt.plot(years, weights, c='r', lw=3, linestyle='--')
```
